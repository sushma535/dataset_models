```python
price=pd.DataFrame(data['price'])
carat=pd.DataFrame(data['carat'])
```

```python
price
```

```python
lm =linear_model.LinearRegression()
model=lm.fit(price,carat)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(price, carat)
```

```python
import numpy as np
price_new = np.array([97])
price_new = price_new.reshape(-1,1)
carat_predict = model.predict(price_new)
carat_predict
```

```python
X=np.array([67,78,94])
X=pd.DataFrame(X)
Y=model.predict(X)
Y=pd.DataFrame(Y)
df=pd.concat([X,Y], axis=1,keys=['price_new','carat_predict'])
df
```

```python
data.plot(kind='scatter', x='price', y='carat')
plt.plot(price,model.predict(price),color='red', linewidth=2)
plt.scatter(price_new, carat_predict, color='black')
plt.plot(df)
plt.show()
```
